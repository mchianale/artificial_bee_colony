# Convergence

```python
from IPython.display import clear_output
from IPython.display import display, HTML
def convergence(f, colony_size, max_iter, limit):
    solutions = []
    for i in range(1, max_iter+1):
        display(HTML(f'<h2>Current Iteration: {i + 1}</h2>'))
        abc = ABC(f, colony_size, i, limit)
        best_answer, best_solution = abc.optimize(displayAns=False)
        solutions.append(best_solution)
        #clear output
        clear_output(wait=True)
    iterations = [i for i in range(1, max_iter+1)]
    fitness = [fit(solution) for solution in solutions]
    plt.figure(figsize=(20, 8))
    plt.plot(iterations, fitness, marker='.',linestyle='')
    plt.xlabel('Iterations')
    plt.ylabel('Best fitness values')
    plt.title('convergence of fitness values according to iterations')
    plt.show()
```

```python
dim = 2
colony_size = 5
limit = 2
max_iter = 2000
f = Rastrigin(dim)
convergence(f, colony_size, max_iter, limit)
```
